```python
import pandas as pd
```

```python
df = pd.read_csv('final2.csv')
```

```python
df.columns
```

```python
test_columns = ['Action', 'Adventure', 'Animation', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Family', 'Fantasy', 'History', 'Horror', 'Music', 'Mystery', 'Romance', 'Science Fiction', 'Thriller', 'War', 'Western', 'Actor_names', 'Director_name', 'original_language', 'year', 'ratings', 'production']
```

```python
len(test_columns)
```

```python
from sklearn.model_selection import train_test_split

X = df[test_columns]
y = df['movie_id']


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2)
```

```python
len(X_test), len(X_train)
```

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier()

model.fit(X_train, y_train)
```

```python
sample_test = X_test.iloc[8]
sample_test
```

```python
indexs = model.kneighbors([sample_test], return_distance= False, )
first_test_index = indexs[0]
first_test_index
```

```python
results = df[df.index.isin(first_test_index)]
results
```
