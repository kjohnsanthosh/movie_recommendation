```python
import pandas as pd
```

```python
df1 = pd.read_csv('movie_actor_director.csv')
df2 = pd.read_csv('movies_details.csv')

df1.head()
```

```python
df2.head()
```

```python
df = pd.merge(df1, df2, on='title', how='inner')
```

```python
df = df.drop('id', axis=1)
```

```python
df.head()
```

```python
df.to_csv('mergetofile.csv', index=False)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['Actor_names'] = le.fit_transform(df['Actor_names'])
df['Director_name'] = le.fit_transform(df['Director_name'])
df['original_language'] = le.fit_transform(df['original_language'])
df['production'] = le.fit_transform(df['production'])
 
```

```python
df.head()
```

```python
df.to_csv('final.csv', index=False)
```
