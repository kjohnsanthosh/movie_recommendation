GENRES FOR MOVIES

```python
df['genres'][0]
```

```python
df['keywords'][0]
```

```python
delete_genres = df[df['genres'] == '[]'].index
df = df.drop(delete_genres)
```

```python
import ast 

genres = []


for data in df['genres']:

    name = ast.literal_eval(data)

    list_of_genres = []


    for val in name:
        val1 = val['name']
        list_of_genres.append(val1)
       

    genres.append(list_of_genres)

df['genres'] = genres
```

```python
df.head()
```

```python
df.drop(['budget', 'keywords'], axis=1, inplace=True)
df.head()
```

```python
df.to_csv('movie_details.csv', index=False)
```

```python
comedy_movies = df[df['genres'].apply(lambda x: 'Crime' in x)]
comedy_movies
```

```python
set_of_genres = {'Action', 'Adventure', 'Animation', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Family', 'Fantasy', 'History', 'Horror', 'Music', 'Mystery', 'Romance', 'Science Fiction', 'Thriller', 'War', 'Western'}
```

```python
def check_for_genre(genres, idx, column):
    if column in genres:
        return idx + 10
    else:
        return 0
    
```

```python
for idx, column in enumerate(set_of_genres):
    df[column] = df["genres"].apply(check_for_genre, args=(idx, column))
```
